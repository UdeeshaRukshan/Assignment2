```python
from langchain_ollama.embeddings import OllamaEmbeddings  # <-- fixed
from langchain_community.vectorstores import Chroma       # <-- fixed
from langchain_community.llms import Ollama                # <-- fixed
from langchain.chains import RetrievalQA
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import PyPDFLoader  # <-- fixed

# 1. Load your PDF document
loader = PyPDFLoader('5secondsrule.pdf')
documents = loader.load()

# 2. Split into smaller chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
docs = text_splitter.split_documents(documents)

# 3. Create embeddings using Ollama + llama3.2
embeddings = OllamaEmbeddings(model="llama3.2:latest")

# 4. Create ChromaDB vector store
db = Chroma.from_documents(docs, embedding=embeddings, persist_directory="./chroma_db")

# Save the database
db.persist()

# 5. Set up retriever
retriever = db.as_retriever()

# 6. Initialize Ollama LLM with llama3.2
llm = Ollama(model="llama3.2:latest")

# 7. Build RetrievalQA chain
qa = RetrievalQA.from_chain_type(llm=llm, retriever=retriever)

# 8. Ask a question
query = "What is the main topic of the document?"
result = qa.run(query)

print("\nAnswer:", result)
```
